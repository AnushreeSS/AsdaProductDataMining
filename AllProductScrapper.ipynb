```python
import import_ipynb
import pandas as pd
from AsdaProductScraper import get_product_info
from time import sleep
from os.path import isfile
```

```python
if isfile("all_urls_tracker.csv"):
    df_urls = pd.read_csv("all_urls_tracker.csv")
    df_urls.drop(df_urls.columns[0], axis=1, inplace=True)
    print("Restored all urls tracker")
else:
    df_urls = pd.read_csv("all_urls.csv")
    df_urls.drop(df_urls.columns[0], axis=1, inplace=True)
    df_urls["scrapped"]=False
    df_urls["success"]=False
df_urls
```

```python
product_data = pd.DataFrame()
if isfile("asda_product_data.csv"):
    product_data = pd.read_csv("asda_product_data.csv")
    product_data.drop(product_data.columns[0], axis=1, inplace=True)
product_data
```

```python
df_urls["scrapped"].value_counts(), df_urls["success"].value_counts()
```

```python
include_titles = ["Life Style", "Allergy Advice", "Additives", "Features", "Recycling Info", "Product Information", 
                      "Brand Marketing", "Other Information", "Third Party Logo"]
for index, row in df_urls[df_urls["scrapped"]==False].iterrows():
    url = row["url"]
    df = get_product_info(url, include_titles)
    df_urls.at[index, "scrapped"] = True
    if not isinstance(df, type(None)):
        df["url_id"] = index
        df_urls.at[index, "success"] = True
        product_data = pd.concat([product_data, pd.DataFrame(df, index=[0])], ignore_index=True)
    if product_data.shape[0]%25==0:
        print(f"Current Size of product data collected: {product_data.shape}")
    sleep(2)
```
